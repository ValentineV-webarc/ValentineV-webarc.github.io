1. **Analisis Penjualan Harian**:
   - Produk mana yang menghasilkan pendapatan tertinggi setiap hari? (Cara 1)


```python
Produk_Tertinggi={'tanggal':[],'produk':[],'Max_Income':[]}
Max=df['pendapatan'][0]
produk=df['produk'][0]
for i in range (0,(len(df)-1)):
    if (df['tanggal'][i]!=df['tanggal'][i+1]):
        Produk_Tertinggi['tanggal'].append(df['tanggal'][i])
        Produk_Tertinggi['produk'].append(produk)
        Produk_Tertinggi['Max_Income'].append(Max)
        Max=df['pendapatan'][i]
        produk=df['produk'][i]
    else:
        if Max < df['pendapatan'][i]:
            Max=df['pendapatan'][i]
            produk=df['produk'][i]

Produk_Tertinggi['tanggal'].append(df['tanggal'].iloc[-1])
Produk_Tertinggi['produk'].append(produk)
Produk_Tertinggi['Max_Income'].append(Max)
Produk_Tertinggi=pd.DataFrame(Produk_Tertinggi)
Produk_Tertinggi
```

2. **Analisis Kategori Produk**:
   - Kategori produk mana yang memiliki jumlah penjualan terbanyak secara keseluruhan?

```python
a=df.groupby('kategori',as_index=False).pendapatan.sum().sort_values(by='pendapatan',ascending=False)
a
```


   - Berapa rata-rata pendapatan per hari untuk setiap kategori produk?

```python
def mean (**kwags):
    data=kwargs[r'dataframe'].copy()
    mean=[]
    for i in range (0,len(data[kwargs['semua']])):
        rata_rata=data[kwargs['semua']][i]//data[kwargs['jumlah']][i]
        mean.append(rata_rata)
    data['rata_rata']=mean
    data=data.groupby(by=['tanggal',kwargs['columns']],as_index=False).agg({'jumlah_terjual':'sum','pendapatan':'sum','rata_rata':'sum'})
    return data
    
mean(dataframe=df,columns='kategori',semua='pendapatan',jumlah='jumlah_terjual')
```

3. **Analisis Harga dan Jumlah Terjual**:

3. **Analisis Harga dan Jumlah Terjual**:
   - Produk mana yang memiliki rasio harga terhadap jumlah terjual tertinggi?

```python
rasio=df['harga_satuan'][0]//df['jumlah_terjual'][0]
for i in range (0,len(df)):
    if rasio < (df['harga_satuan'][i]//df['jumlah_terjual'][i]):
        rasio=df['harga_satuan'][i]//df['jumlah_terjual'][i]
        indeks=i
        
print(df.loc[indeks])
print("rasio:",rasio)
```

4. **Tren Penjualan Produk**:
   - Bagaimana tren penjualan (jumlah terjual) dari produk Laptop A selama enam hari?

```python
Mac = df[df['produk']=="Laptop MacBook"]
data = [Mac]
title = ["Laptop MacBook"]
indeks = 0
for i in data:
    plt.figure(figsize=(10,3))
    sns.lineplot(x="tanggal",y="jumlah_terjual",data=i)
    plt.show()
    print("\n",f"{title[indeks]:>23}\n")
    print(i.groupby(by="tanggal",as_index=False).jumlah_terjual.sum())
    indeks+=1
    
```
