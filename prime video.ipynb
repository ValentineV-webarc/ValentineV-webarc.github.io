```python
data=df['Username'].value_counts().reset_index()
loyal_custumer={
    'nama' : [],
    'count' : []
}
for i in range(0,len(data['count'][:])):
    if data['count'][i]>1:
        loyal_custumer['nama'].append(data['Username'][i])
        loyal_custumer['count'].append(data['count'][i])
```

```python
for i in range(0,len(loyal_custumer['count'][:])):
    for j in loyal_custumer:
        print ("{:<20}".format(loyal_custumer[j][i]),end = " ") #format utk loyal_customer masuk berjarak
    print()
```
